# Empirical KL Divergence

```python
ball_para = ball[['Mean Curvature','Mean Angle']]
convex_para = convex[['Mean Curvature','Mean Angle']]
freeform_para = freeform[['Mean Curvature','Mean Angle']]
```

```python
curvature_step = np.linspace(start=0,stop=0.4,num=30)
angle_step = np.linspace(start=-1,stop=1,num=30)
```

```python
def cal_dist(data, curvature_step, angle_step):
    num_data = len(data)
    dist = []
    for i in range(len(curvature_step)-1):
        for j in range(len(angle_step)-1):
            flag = (data['Mean Curvature'] > curvature_step[i]) & (data['Mean Curvature'] <= curvature_step[i+1]) & (data['Mean Angle'] > angle_step[i]) & (data['Mean Angle'] < angle_step[i+1])
            freq = len(data[flag])
            dist.append(freq / num_data)
    dist = np.array(dist)
    return(dist)
```

```python
ball_dist = cal_dist(ball_para,curvature_step,angle_step)
convex_dist = cal_dist(convex_para,curvature_step,angle_step)
freeform_dist = cal_dist(freeform_para,curvature_step,angle_step)
```

```python
def cal_dist_2(data, curvature_step, angle_step):
    num_data = len(data)
    dist = []
    for i in range(len(curvature_step)-1):
        for j in range(len(angle_step)-1):
            flag = (data['curvature'] > curvature_step[i]) & (data['curvature'] <= curvature_step[i+1]) & (data['angle'] > angle_step[i]) & (data['angle'] < angle_step[i+1])
            freq = len(data[flag])
            dist.append(freq / num_data)
    dist = np.array(dist)
    return(dist)
```

```python
test_dist = cal_dist_2(test_d[2],curvature_step,angle_step)
```

```python
def kullback_leibler_divergence(p, q):
    sum_val = 0
    for i in range(len(q)):
        if q[i] == 0:
            continue
        else:
            if p[i] == 0:
                sum_val +=-1000*q[i]
            else:
                sum_val += p[i] * np.log(p[i]/q[i])
                
    return sum_val
```

```python
# freeform vs others
print(kullback_leibler_divergence(ball_dist,test_dist))
print(kullback_leibler_divergence(convex_dist,test_dist))
print(kullback_leibler_divergence(freeform_dist,test_dist))
```

```python
kld = pd.DataFrame()
for num in np.linspace(5,40,num=8):
    num = np.int(num)
    curvature_step = np.linspace(start=0,stop=0.4,num=num)
    angle_step = np.linspace(start=-1,stop=1,num=num)
    
    ball_dist = cal_dist(ball_para,curvature_step,angle_step)
    convex_dist = cal_dist(convex_para,curvature_step,angle_step)
    freeform_dist = cal_dist(freeform_para,curvature_step,angle_step)
    
    test_dist = cal_dist_2(test_d[2],curvature_step,angle_step)
    
    ball_kld = kullback_leibler_divergence(ball_dist,test_dist)
    convex_kld = kullback_leibler_divergence(convex_dist,test_dist)
    freeform_kld = kullback_leibler_divergence(freeform_dist,test_dist)
    
    kld = kld.append(pd.DataFrame({'bin_num': [num], 'kld': [ball_kld], "Object": ["Half Ball"]}))
    kld = kld.append(pd.DataFrame({'bin_num': [num], 'kld': [convex_kld], "Object": ["Freeform 1"]}))
    kld = kld.append(pd.DataFrame({'bin_num': [num], 'kld': [freeform_kld], "Object": ["Freeform 2"]}))
```

```python
kld.reset_index(drop=True, inplace=True)
```

```python
plt.figure(figsize=(8,6), dpi=300)
plt2 = sns.lineplot(x="bin_num", y="kld", hue="Object", marker='o', data=kld, linewidth=3, markersize=10)
plt2.xaxis.set_label_text('Number of Bins $l$',fontsize=25);
plt2.yaxis.set_label_text('Empirical KL Divergence',fontsize=25);
plt.legend(bbox_to_anchor=(1.04,1), fontsize=25);
plt.xticks(np.arange(5, 45.1, 10), fontsize=25);
plt.yticks(np.arange(-750, 50.1, 200), fontsize=25);
```
